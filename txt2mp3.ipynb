# Task
Convert text files to speech (MP3) using `gTTS`.

## Listar archivos de texto

### Subtask:
Identify all text files (e.g., with a `.txt` extension) in the specified folder.


**Reasoning**:
First, I need to import the necessary modules, `os` for operating system interactions and `glob` for pattern matching, as specified in the instructions.



````python
import os
import glob

print("Imported os and glob modules.")
````

**Reasoning**:
Now that the necessary modules are imported, I will define a default input folder path. Since there are no `.txt` files in the provided file list, I will create a dummy `.txt` file in the specified input folder to ensure the `glob` function has something to identify and demonstrate its functionality. Then, I will use `glob.glob()` to find all `.txt` files within that folder and print the list.



````python
input_folder = '/content/sample_data/'

# Create a dummy .txt file for demonstration purposes
dummy_file_path = os.path.join(input_folder, 'dummy_text_file.txt')
with open(dummy_file_path, 'w') as f:
    f.write('This is a dummy text file.\n')

print(f"Created a dummy text file at: {dummy_file_path}")

# Use glob to find all .txt files in the input_folder
text_files = glob.glob(os.path.join(input_folder, '*.txt'))

print(f"Identified text files: {text_files}")
````

## Leer contenido de cada archivo

### Subtask:
Leer contenido de cada archivo de texto identificado.


**Reasoning**:
I will iterate through the `text_files` list, open each file, read its content using UTF-8 encoding, and print both the filename and its content as requested by the instructions.



````python
for file_path in text_files:
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
        print(f"\n--- Content of {file_path} ---")
        print(content)
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
````

## Convertir texto a voz (MP3)

### Subtask:
Utilizar la biblioteca `gTTS` para convertir el contenido de cada archivo de texto en un objeto de audio de voz.


**Reasoning**:
The first step according to the instructions is to install the `gTTS` library. This is typically done using `pip` in a separate command.



````python
!pip install gTTS
print("gTTS library installed.")
````

**Reasoning**:
Now that `gTTS` is installed, I will import the `gTTS` class, create an empty list for speech objects, and then iterate through each text file. For each file, I will read its content, create a `gTTS` object with the content, and append it to the `speech_objects` list as per the instructions.



````python
from gtts import gTTS

speech_objects = []

for file_path in text_files:
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            text_content = f.read()

        # Create a gTTS object. Using Spanish from Spain ('es', tld='es') and setting slow=False for faster speech.
        # The content of the dummy file is in English, so the output might sound a bit off for this specific text.
        # For actual Spanish text, this setting will work perfectly.
        tts = gTTS(text=text_content, lang='es', tld='es', slow=False)
        speech_objects.append(tts)
        print(f"Created gTTS object for '{file_path}' with Spanish (Spain) voice and faster speed")
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

print(f"Total speech objects created: {len(speech_objects)}")
````

## Guardar archivos MP3

### Subtask:
Guardar cada archivo MP3 generado en una carpeta de salida, con un nombre que corresponda al archivo de texto original.


**Reasoning**:
I will define the output folder, create it if it doesn't exist, and then iterate through the `speech_objects` and corresponding `text_files` to save each `gTTS` object as an MP3 file with a derived filename in the output folder, printing a confirmation for each.



````python
output_folder = '/content/audio_output/'
os.makedirs(output_folder, exist_ok=True)
print(f"Output folder '{output_folder}' ensured to exist.")

for i, tts_object in enumerate(speech_objects):
    try:
        original_file_path = text_files[i]
        base_filename = os.path.basename(original_file_path)
        filename_without_ext = os.path.splitext(base_filename)[0]
        mp3_filename = f"{filename_without_ext}.mp3"
        output_file_path = os.path.join(output_folder, mp3_filename)

        tts_object.save(output_file_path)
        print(f"Successfully saved '{original_file_path}' as '{output_file_path}'")
    except Exception as e:
        print(f"Error saving MP3 for '{text_files[i]}': {e}")
````

#### Análisis de la velocidad de `gTTS`:

Como se mencionó previamente en el código, al usar `gTTS(slow=False)`, ya estamos solicitando la máxima velocidad de habla que la biblioteca `gTTS` puede ofrecer por defecto. Si esta velocidad aún no es suficiente, necesitamos considerar enfoques alternativos.

#### Alternativas para una mayor velocidad de habla:

Existen varias bibliotecas y APIs de conversión de texto a voz que ofrecen mayor control sobre la velocidad de habla, a menudo a través de parámetros específicos o la capacidad de ajustar el "rate" o "pitch" de la voz. Algunas opciones notables incluyen:

1.  **Google Cloud Text-to-Speech API**: Esta es una API de pago que ofrece voces de alta calidad y un control granular sobre varios parámetros del habla, incluyendo la velocidad. Permite ajustar el `speaking_rate` en un rango que va desde 0.25 (muy lento) hasta 4.0 (muy rápido), siendo 1.0 la velocidad normal.
2.  **Amazon Polly**: Otra API de pago que proporciona voces realistas y un control extenso sobre la pronunciación, el tono y la velocidad del habla a través de SSML (Speech Synthesis Markup Language). Polly permite especificar la velocidad del habla usando el elemento `<prosody rate="x%">` donde `x` puede ser un porcentaje de la velocidad normal.
3.  **IBM Watson Text to Speech**: Ofrece voces de alta calidad con la capacidad de personalizar la velocidad y el tono. También soporta SSML para un control detallado sobre la salida de voz.
4.  **Microsoft Azure Text to Speech**: Similar a las anteriores, esta API de Microsoft Azure ofrece voces neuronales y un amplio control sobre la voz, el estilo y la velocidad a través de SSML.
5.  **`pyttsx3`**: Una biblioteca de Python que funciona offline y utiliza los motores de síntesis de voz instalados en el sistema operativo (e.g., SAPI en Windows, NSSpeechSynthesizer en macOS, espeak en Linux). Permite ajustar la propiedad `rate` para cambiar la velocidad del habla.

#### Métodos de post-procesamiento de audio:

Si no se desea cambiar la biblioteca de TTS, otra opción es generar el MP3 con `gTTS` y luego post-procesar el archivo de audio para acelerar su reproducción. Esto se puede hacer utilizando:

1.  **Software de edición de audio**: Herramientas como Audacity, Adobe Audition, o cualquier DAW (Digital Audio Workstation) permiten importar un archivo MP3 y ajustar su velocidad de reproducción sin cambiar el tono.
2.  **Bibliotecas de Python para procesamiento de audio**: Bibliotecas como `pydub` o `librosa` en combinación con `ffmpeg` pueden ser usadas para manipular la velocidad de un archivo MP3. Por ejemplo, `pydub` puede cambiar la velocidad de un segmento de audio. Un ejemplo de uso de `pydub` para acelerar un audio sería:
    ```python
    from pydub import AudioSegment
    audio = AudioSegment.from_mp3("input.mp3")
    fast_audio = audio.speedup(playback_speed=1.5) # Acelera 1.5 veces
    fast_audio.export("output_fast.mp3", format="mp3")
    ```
